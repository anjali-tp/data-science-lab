```python


import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 12))

# Graph 1
x = np.array([1, 3])
y = np.array([4, 7])

plt.subplot(3, 3, 1)
plt.title('Graph 1')
plt.xlabel("X axis", color="green")
plt.ylabel("Y axis", color="green")
plt.plot(x, y, 's', color="green")

# Graph 2
y1 = np.array([2, 7, 3, 10])
y2 = np.array([5, 1, 9, 4])

plt.subplot(3, 3, 2)
plt.plot(y1, color="purple", marker='x')
plt.plot(y2, linewidth=2, color="red", linestyle="dotted")

# Graph 3
ypoints = np.array([2, 9, 4, 12])

plt.subplot(3, 3, 3)
plt.plot(ypoints, marker='s', ms=12, mec="blue", mfc="green")

# Graph 4
x = np.array([70, 75, 80, 85, 90, 95, 100, 105, 110, 115])
y = np.array([200, 210, 220, 230, 240, 250, 260, 270, 280, 290])

plt.subplot(3, 3, 4)
plt.title("Graph 2")
plt.plot(x, y)
plt.grid()

# Graph 5
x = np.random.randint(30, size=(50))
y = np.random.randint(30, size=(50))
colors = np.random.randint(30, size=(50))
sizes = 5 * np.random.randint(30, size=(50))

plt.subplot(3, 3, 5)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')

# Graph 6
x = np.array(["W", "X", "Y", "Z"])
y = np.array([4, 9, 2, 11])

plt.subplot(3, 3, 6)
plt.bar(x, y, color="cyan", width=0.2)

# Graph 7 - Pie Chart
plt.subplot(3, 3, 7)
y = np.array([30, 20, 30, 20])
mylabels = ["Pears", "Plums", "Peaches", "Nectarines"]
myexplode = [0.1, 0, 0, 0]
plt.pie(y, labels=mylabels, explode=myexplode, startangle=120, shadow=True, autopct='%1.1f%%')
plt.legend(title="Fruit Types:", bbox_to_anchor=(1.05, 1), loc='upper left')

# Graph 8 - Simple Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.subplot(3, 3, 8)
plt.plot(x, y, color='magenta')
plt.title("Simple Line Plot")

# Graph 9 - Simple Bar Chart
x = ['A', 'B', 'C', 'D']
y = [5, 7, 3, 9]

plt.subplot(3, 3, 9)
plt.bar(x, y, color='orange')
plt.title("Simple Bar Chart")

plt.suptitle("All Graphs", fontsize=15)

plt.subplots_adjust(left=0.1, right=0.8, top=0.9, bottom=0.1, wspace=0.3, hspace=0.3)

plt.show()

```
